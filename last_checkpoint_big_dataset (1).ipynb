```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules 
import csv
```

```python
with open('Market_Basket_Optimisation.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
```

```python
te=TransactionEncoder()
te_ary=te.fit(data).transform(data)
df=pd.DataFrame(te_ary, columns=te.columns_) 
df.head()
```

```python
encoder = LabelEncoder()
for i in df : 
  df[i] = encoder.fit_transform(df[i])
df.head()
```

```python
frequent_itemsets=apriori(df, min_support=0.1, use_colnames=True) # it's true that 0.1 is considered as a low value but i selected it jsut to seee the items with higher suport in df
frequent_itemsets
```

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.6)
```

```python
association_rules(frequent_itemsets,metric="lift",min_threshold=1) # I selected 1 as min_threshhold because any value under one the two items are unlikely to be bought together
```
